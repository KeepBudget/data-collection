# Property 매매
[서울 열린데이터 광장 - 서울시 부동산 실거래가 정보](https://data.seoul.go.kr/dataList/OA-21275/S/1/datasetView.do)

- id - auto_increment
- district_id - 자치구코드
- address - 자치구명 법정동명 본번 부번
- name - 건물명
- floor - 층
- price - 물건금액(만원)
- size - 건물면적(m^2) / 3.3
- type - 건물용도
- trade_type - DEAL

```python
import pandas as pd
```

```python
# column 다 보이게
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('data/seoul-deal-2024.csv', encoding='EUC-KR')
df.head()
```

```python
df.columns
```

```python
# 1차적으로 필요없는 column 삭제
df.drop(['접수연도', '법정동코드', '지번구분', '지번구분명', '계약일', '토지면적(㎡)', '권리구분', '취소일', '건축년도', '신고구분', '신고한 개업공인중개사 시군구명'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# 결측치 제거
df = df.dropna()
```

```python
# address 추출
df['address'] = df['자치구명'] + " " + df['법정동명'] + " " +  df['본번'].astype(int).astype('string') + "-" + df['부번'].astype(int).astype('string')
```

```python
df.head()
```

```python
# column명 변경 & 필요 없는 column 제거
df.rename(columns={'자치구코드':'district_id', '건물명':'building_name', '물건금액(만원)':'price', '건물면적(㎡)':'size', '층':'floor', '건물용도':'type'}, inplace=True)
df.drop(['자치구명', '법정동명', '본번', '부번'], axis=1, inplace=True)
```

```python
df['trade_type']='DEAL'
```

```python
df.head()
```

```python
# 면적 -> 평수
df['size'] = df['size'].astype('float') / 3.3
```

```python
df['floor'] = df['floor'].astype('int')
```

```python
df.head()
```

```python
df['type'].value_counts()
```

```python
def changeTypeName(type):
  if type == '아파트':
    return 'APARTMENT'
  elif type == '연립다세대':
    return 'MULTIPLEX_HOUSING'
  elif type == '오피스텔':
    return 'OFFICETEL'
```

```python
df['type'] = df['type'].apply(changeTypeName)
```

```python
df.head()
```

```python
df= df.loc[~df['building_name'].str.startswith('(')]
```

```python
df.head()
```

## 데이터 DB에 넣기

```python
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

username= os.getenv("DB_USERNAME")
password = os.getenv("DB_PASSWORD")
host = os.getenv("DB_HOST")
port = os.getenv("DB_PORT")
database = os.getenv("DB_DATABASE")

engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')
```

```python
tableName = 'properties'
```

```python
df.to_sql(name=tableName, con=engine, if_exists='append', index=False)
```
